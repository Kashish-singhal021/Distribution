Q2. Generate 100 values in Excel using the continuous uniform distribution RAND() and plot a histogram.
Describe the shape of the distribution.

                 - Step 1: Generate 100 random values

In cell A1, type:

=RAND()

Drag the formula down to A100.

This generates 100 random numbers between 0 and 1.

RAND() produces a continuous uniform distribution.

Step 2: Create a histogram

Select the 100 values (A1:A100).

Go to Insert → Charts → Histogram (or Insert → Statistics Chart → Histogram depending on Excel version).

Excel will automatically group the values into bins.

Optional:

You can adjust bin width for a smoother histogram:

Right-click the horizontal axis → Format Axis → Bin width (e.g., 0.1)

Step 3: Describe the shape

The histogram will show approximately equal heights across bins.

Shape: Rectangular / Uniform

Each value is equally likely in the interval [0,1]

There are no peaks or skewness

Step 4: Notes

With more values (e.g., 1000), the histogram will look even flatter, closer to a perfect uniform distribution.

Unlike normal distributions, uniform distributions have constant probability density across the interval.

Q4. What is the concept of standardization (z-score), and why is it important in data analysis? Explain the
formula and how standardization transforms a dataset.

             - 1️⃣ Concept of Standardization (Z-score)

Standardization is the process of transforming individual data points into a common scale with a mean of 0 and a standard deviation of 1.

This is done by calculating the z-score for each value.

The z-score tells you how many standard deviations a value is above or below the mean.

Purpose:

Makes variables comparable, even if they are measured in different units (e.g., height in cm, weight in kg).

Useful in statistical analysis, machine learning, and hypothesis testing.

Q5. What is Kurtosis and their type?

          - 1️⃣ Concept of Kurtosis

Kurtosis measures the “tailedness” or peakedness of a probability distribution.

It tells us how heavy or light the tails of a distribution are compared to a normal distribution.

High kurtosis → more data in the tails → more extreme values (outliers)

Low kurtosis → less data in tails → flatter distribution

Note: Kurtosis focuses on the shape of the distribution, not its spread (variance) or center (mean).

Characteristics: Flatter peak, fewer extreme values

Example: Uniform distribution

Q8. How does log transformation help in stabilizing variance and making data more normally distributed?

               - 1️⃣ Concept of Log Transformation

A log transformation applies the logarithm (usually natural log ln or base 10 log) to each data point:

Purpose:

Reduces skewness in the data

Stabilizes variance, especially when the variance increases with the mean (heteroscedasticity)

2️⃣ How Log Transformation Stabilizes Variance

In many datasets, larger values have larger variability (heteroscedasticity).

Taking the log compresses large values more than small values.

This reduces the range of variation and makes variance more uniform across the data.

Example:

Original X	Log(X)
10	1
100	2
1000	3

Variance of original data is large because 1000 is far from 10.

After log transformation, the spread is much smaller → variance is stabilized.

3️⃣ How Log Transformation Makes Data More Normal

Many real-world datasets (like income, population, or reaction times) are positively skewed.

Applying log transformation:

Compresses high values

Expands low values slightly

Brings the distribution closer to symmetry

Effect:

Data becomes approximately normally distributed, which is useful for:

t-tests, ANOVA

Regression

Other parametric statistical methods

4️⃣ Summary
Benefit	Explanation
Variance stabilization	Reduces the impact of large values; variance becomes more uniform
Skewness reduction	Compresses high values; distribution becomes more symmetric
Normality approximation	Makes data suitable for parametric statistical methods
